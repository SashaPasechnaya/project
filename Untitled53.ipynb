```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import time
```

Загрузка датасета LFW

```python

lfw_dataset = fetch_lfw_people(min_faces_per_person=100, resize=0.4)

images = lfw_dataset.images
labels = lfw_dataset.target
target_names = lfw_dataset.target_names

print("Размер массива изображений:", images.shape)
print("Количество меток:", len(labels))
print("Количество классов:", len(target_names))
print("Список имен классов:", target_names)
```

```python
for i in range(5):
    plt.figure(figsize=(2, 2))
    plt.imshow(images[i], cmap='gray')
    plt.title(target_names[labels[i]])
    plt.axis('off')
    plt.show()
```

```python
first_image_indices = {}

# Поиск индексов первых изображений каждого класса
for i, label in enumerate(labels):
    if label not in first_image_indices:
        first_image_indices[label] = i

# Вывод по одному изображению для каждого класса
for label, index in first_image_indices.items():
    plt.figure(figsize=(2, 2))
    plt.imshow(images[index], cmap='gray')
    plt.title(target_names[label])
    plt.axis('off')
    plt.show()
```

Нормалізація до (-1, 1) та зведення міток до категоріальних

```python
images = ((images/255.) - .5)*2
labels_cat = to_categorical(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels_cat, test_size=0.2, random_state=42)
```

Створимо модель с одним вихідним шаром softmax

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten()) 
model.add(tf.keras.layers.Dense(5, activation = 'softmax',kernel_initializer=tf.keras.initializers.RandomNormal(mean=0.0, stddev=1.0) ))
```

```python
model.compile(optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01), loss = 'categorical_crossentropy',
             metrics = ['accuracy', 'categorical_accuracy', tf.keras.metrics.Precision(),
                        tf.keras.metrics.Recall()])
```

```python
start_time = time.time()
hist = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 100, batch_size = 30, verbose = 1)
end_time = time.time()
total_time = end_time - start_time
print("Время обучения:", total_time, "сек")
```

```python
history = hist.history
fig = plt.figure(figsize = (16, 4))
ax = fig.add_subplot(1, 2, 1)
plt.plot(history['loss'], lw = 4)
plt.plot(history['val_loss'], lw = 4)
plt.legend(['Потеря при обучении', 'Потеря при проверке'], fontsize = 15)
ax.set_xlabel('Эпохи', size = 15)
ax = fig.add_subplot(1, 2, 2)
plt.plot(history['accuracy'], lw = 4)
plt.plot(history['val_accuracy'], lw = 4)
plt.legend(['Правильность при обучении', 'Правильность при  проверке'], fontsize = 15)
ax.set_xlabel('Эпохи', size = 15)
```

Так як наші дані це зображення, можемо спробувати алгоритм Adagrad, бо він гарно працює з розрідженими даними. Як активацію виберемо ELU з міркувань, що вона має згладжений негативний нахил і експоненційне зростання для позитивних значень.
Може допомогти у зниженні зміщення (bias) та прискоренні навчання.
Зазвичай працює добре на задачах із розрідженими даними.

```python
NN3 = tf.keras.Sequential()
NN3.add(tf.keras.layers.Flatten())
NN3.add(tf.keras.layers.Dense(64, activation='elu',
                             kernel_initializer=tf.keras.initializers.HeNormal()))
NN3.add(tf.keras.layers.Dense(64, activation='elu'))

NN3.add(tf.keras.layers.Dense(5, activation='softmax'))
```

```python
NN3.compile(optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.001), loss = 'categorical_crossentropy',
             metrics = ['accuracy', 'categorical_accuracy', tf.keras.metrics.Precision(),
                        tf.keras.metrics.Recall()])
```

```python
start_time = time.time()
history4 = NN3.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 100, batch_size = 20, verbose = 1)
end_time = time.time()
total_time = end_time - start_time
print("Время обучения:", total_time, "сек")
```

Висновок такий, що найменш затратний по часу та дуже ефективний алгоритм для моїх даних це Adagrad, функція активації ELU та нормалізація за мiнi-батчами перед кожним шаром це найкращий вибір.

Бачимо, що з додаванням нормалiзацiї за мiнi-батчами перед кожним шаром точність покращується, можна вибирати більший learning_rate, перенавчання немає.

```python
NN5 = tf.keras.Sequential()
NN5.add(tf.keras.layers.Flatten())

# Нормализация за мини-батчами перед первым слоем Dense
NN5.add(tf.keras.layers.BatchNormalization())

NN5.add(tf.keras.layers.Dense(120, activation='elu',
                              kernel_initializer=tf.keras.initializers.HeNormal()))
NN5.add(tf.keras.layers.Dropout(0.3))
# Нормализация за мини-батчами перед вторым слоем Dense
NN5.add(tf.keras.layers.BatchNormalization())

NN5.add(tf.keras.layers.Dense(64, activation='elu'))

NN5.add(tf.keras.layers.Dense(5, activation='softmax'))
```

```python
NN5.compile(optimizer = tf.keras.optimizers.Adagrad(learning_rate=0.01), loss = 'categorical_crossentropy',
             metrics = ['accuracy', 'categorical_accuracy', tf.keras.metrics.Precision(),
                        tf.keras.metrics.Recall()])
```
